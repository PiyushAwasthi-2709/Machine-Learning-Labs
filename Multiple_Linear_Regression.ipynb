```python
# importing the libraries
import numpy as np

import copy , math

import matplotlib as plt

import pandas as pd
```

```python
# loading the data
df = pd.read_csv("house_prices.csv")
```

```python
# splitting the features and target

X = df[['Size', 'Bedrooms', 'Age']] 

y = df['Price']
```

```python
X.shape # -> (20,3) -> 20 training examples and 3 features 
```

```python
y.shape # -> (20,) -> Target Output
```

```python
# train_test_split 
def train_test_split(X,y,test_size=0.2):

  m = X.shape[0]

  test_samples = int(test_size * m)

  split_index = m - test_samples

  X_train = X.iloc[:split_index]

  X_test = X.iloc[split_index:]

  y_train = y.iloc[:split_index]
  
  y_test = y.iloc[split_index:]
  

  return X_train,X_test,y_train,y_test

  
```

```python
# normalization
def z_score_normalize(X):

  mu = np.mean(X,axis=0) # mean

  sigma = np.std(X,axis=0) # standard deviation
  
  X_norm = (X - mu) / sigma

  return X_norm ,mu ,sigma
```

```python
# splitting

X_train,X_test,y_train,y_test = train_test_split(X,y)

# normalize 

X_train, mu , sigma = z_score_normalize(X_train)

X_test = (X_test - mu) / sigma

# conversion to numpy arrays 

X_train = X_train.to_numpy()

X_test = X_test.to_numpy()

y_train = y_train.to_numpy()

y_test = y_test.to_numpy() 
```

```python
# initializing parameter -> setting w and b to be zeroes

def initialize_params(n):

  w = np.zeros(n)
  
  b = 0 

  return w,b
```

```python
# prediction function 
def prediction(X,w,b):
  
  y_wb = np.dot(X,w) + b 

  return y_wb
```

```python
# test - 1
w,b = initialize_params(X.shape[1])

y_pred = prediction(X_train,w,b)

print(y_pred) # -> full of zeroes

print(y_pred.shape) # -> taking 16 x_train examples 
```

```python
# computing cost 

def cost_function(X_train,y_train,w,b):
  
  m = y_train.shape[0]

  cost = 0.0

  for i in range(m):
    f_wb_i = np.dot(X_train[i],w) + b 
    cost = cost + (f_wb_i - y_train[i])**2
  total_cost =  cost / (2 * m) 

  return total_cost
```

```python
# compuiting gradient 

def compute_gradient(X_train,y_train,w,b):

  m,n  = X_train.shape

  dj_dw = np.zeros((n,))

  dj_db = 0. 

  for i in range(m):

    err = (np.dot(X_train[i],w) + b) - y_train[i]

    for j in range(n):

      dj_dw[j] = dj_dw[j] + err * X_train[i,j]

    dj_db = dj_db + err

  dj_dw = dj_dw / m

  dj_db = dj_db  / m

  return dj_dw,dj_db
```

```python
# gradient_descent 

def gradient_descnet(X_train , y_train , w , b , cost_function , compute_gradient , alpha , iters):

    J_history = []
    w_in = copy.deepcopy(w)  #avoid modifying global w within function
    b_in = b

    for i in range(iters):
        
        dj_dw , dj_db = compute_gradient(X_train,y_train,w_in,b_in)

        w_in = w_in - alpha * dj_dw
        b_in = b_in - alpha * dj_db

        # saving cost J at each iteration 

        if i < 100000:  #prevent resource exhuastion 

            J_history.append(cost_function(X_train,y_train,w_in,b_in))

        # printing cost at each intervals 10 times or as many iterations if < 10 

        if i % math.ceil(iters/10) == 0:
             
             print(f"Iteration {i:4d}: Cost {J_history[-1]:8.2f}   ")

    return J_history , w_in , b_in
```

```python
# making a small wrapper to fir the training examples 

def fit(X_train,y_train):
   
  #  unpacking w and b

   w , b = initialize_params(X_train.shape[1]) 

   iters = 1000

   alpha = 0.0029
   
   j_hist,w_final,b_final= gradient_descnet(X_train,y_train,w,b,cost_function,compute_gradient,alpha,iters)

   m,_ = X_train.shape
   for i in range(m):
      print(f"prediction: {np.dot(X_train[i], w_final) + b_final:0.2f}, target value: {y_train[i]}")


   return j_hist,w_final,b_final
```

```python
j_hist , w , b = fit(X_train, y_train)
```

```python
# Predictions on training data
y_pred = prediction(X_train, w , b)

plt.figure(figsize=(8, 6))
plt.scatter(y_train, y_pred)

# Perfect prediction line
min_val = min(y_train.min(), y_pred.min())
max_val = max(y_train.max(), y_pred.max())

plt.plot([min_val, max_val],
         [min_val, max_val])

plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted House Prices")

plt.show()
```
