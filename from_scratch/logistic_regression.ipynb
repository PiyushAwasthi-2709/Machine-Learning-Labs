```python
# Implemeting the Sigmoid function : change X_1 to X_train after cheking the test below 
def sigmoid_function(z):

  return 1 / (1 + np.exp(-z))
```

```python
# Implemeting the Cost  function : change X_1 to X_train after cheking the test below and same for y_1 as y_train 
def compute_logistic_cost(X_1,y_1,w,b):

  # assuming sigmoid f_wb to p 
  z = np.dot(X_1,w) + b 

  p = sigmoid_function(z)

  m = X_1.shape[0]

  # calculating total cost 

  cost = np.sum((( y_1 * np.log(p)) + (( 1 - y_1 ) * np.log(1 - p))))

  total_cost = -(cost / m)

  return total_cost
```

```python
def compute_logistic_gradient(X_1,y_1,w,b):

  m , n = X_1.shape

  dj_dw = np.zeros(n)
  dj_db = 0

  for i in range(m):
    z = np.dot(X_1[i],w) + b 

    error = sigmoid_function(z) - y_1[i]

    for j in range(n):
      dj_dw[j] += error * X_1[i,j]
    dj_db += error

  dj_dw = dj_dw / m    
  dj_db = dj_db / m    

  return dj_dw, dj_db
```

```python
def logistic_gradient_descent(X_1 , y_1 , w , b , alpha , iters ) :

  
  J_history = []

  w_in = copy.deepcopy(w)

  b_in = b

  for i in range(iters):
    
    
    dj_dw , dj_db = compute_logistic_gradient(X_1 , y_1 , w_in , b_in)

    # updating the params 

    w_in = w_in - alpha * dj_dw
    b_in = b_in - alpha * dj_db

    # Save cost J at each iteration
    if i<100000:      # prevent resource exhaustion 
      J_history.append( compute_logistic_cost(X_1, y_1, w_in, b_in) )

    # Print cost every at intervals 10 times or as many iterations if < 10
    if i% math.ceil(iters / 10) == 0:
      print(f"Iteration {i:4d}: Cost {J_history[-1]}   ")
        
  return w_in, b_in, J_history
```
