```python
# Full Pipeline of Linear Regression with one Variable 

import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# calculating prediction 
def cal_prediction(x_train,w,b):
    
    # initalizing values of w and b 
    m = x_train.shape[0] # getting the taining examples
    
    f_wb = np.zeros(m) #storing the predficted values 
    
    for i in range(m):
        
        f_wb[i] = w * x_train[i] + b 
        
    return f_wb
```

```python
# calucating the errors
def compute_errors(f_wb, y_train):
    
    m = y_train.shape[0]
    
    error = np.zeros(m)
    
    for i in range(m):
        error[i] = f_wb[i] - y_train[i]
        
    return error
```

```python
def compute_cost(error):
    
    m = error.shape[0]
    
    cost = 0 
    
    for i in range(m):
        
        cost = cost + (error[i])**2 

    total_cost = 1 / (2*m) * cost

    return total_cost

        
```

```python
#implementing Gradient descent 
def compute_gradient(error,x_train):
    
    m = error.shape[0] 
    
    df_dw = 0
    
    df_db = 0
    
    for i in range(m):
        
        df_dw = df_dw + (error[i] * x_train[i])
        
        df_db = df_db + error[i] 
        
    df_dw = (1/m) * df_dw
    
    df_db = (1/m) * df_db

    return df_dw , df_db
```

```python
# implementing function for updating parameters 
def compute_batch_gradient(df_dw,df_db,alpha,w,b):

    # updating parameters 
    w = w - (alpha*df_dw)
    b = b - (alpha*df_db)

    return w,b
```

```python
def Linear_Regression(x_train,y_train,w,b,alpha,num_iters):
    
    cost_history =[]  #storing cost 
    
    for i in range(num_iters):

    # step-1 predictions

        f_wb = cal_prediction(x_train,w,b)

    #step-2 calculating errors 

        err = compute_errors(f_wb,y_train)

    #step-3 computing cost 

        cost = compute_cost(err)

    # storing cost 
        cost_history.append(cost)

    # step-4 gradients 

        df_dw , df_db = compute_gradient(err,x_train)

    # step-5 Updating parameters

        w,b = compute_batch_gradient(df_dw, df_db, alpha, w, b)

        # print progress
        print(f"Iteration {i}: Cost = {cost:.4f}, w = {w:.4f}, b = {b:.4f}")

    return  w , b , cost_history
        
        
```
