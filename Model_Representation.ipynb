```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# x_train is the input variable (size in 1000 square feet)
# y_train is the target (price in 1000s of dollars)
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
# m -> training examples
# array.shape -> returns a tuple with an entry for each dimesion x_train.shape[0] -> length of array 
print(f"x_train.shape : {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples is: {m}")
```

```python
# One can also use the Python len() function as shown below.
# m is the number of training examples
m = len(x_train)
print(f"Number of training examples is: {m}")
```

```python
# training examples -> ith training examples x_i , y_i
i = 1 # change this to 1 to see x^1 , y^1 
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^{i},y^{i}) :({x_i},{y_i})")
```

```python
# You can plot these two points using the scatter() function in the matplotlib library, as shown in the cell below.

# The function arguments marker and c show the points as red crosses (the default is blue dots).
# You can use other functions in the matplotlib library to set the title and labels to display

#time to plot babyyyy!!!

plt.scatter(x_train,y_train,marker='.',c='b')
# labels for x , y and the plot itself 
# for the plot 
plt.title("Housing Prices")
# label for x 
plt.xlabel("Size (1000 sqft)")
# label for y
plt.ylabel("Price (in 1000s of dollars)")
plt.show()
```

```python
# model computation 

def compute_model_output(x,w,b):
    m = x.shape[0]
    f_wb = np.zeros(m) # making an array of zeros 
    for i in range(m):
        f_wb[i] = w * x[i] + b  # storing the new predicted values in place of those zeros 

    return f_wb 
```
